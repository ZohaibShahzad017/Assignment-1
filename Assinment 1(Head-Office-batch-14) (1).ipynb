# 🔹 Q12 (Full Name + Upper/Lower)  User se first name aur last name lo aur:  1)  full name print karo  2) upper case 3)lower case

```python
# type your code here
```

```python

first_name = input("Enter first name: ")
last_name = input("Enter last name: ")

full_name = first_name + " " + last_name

print("Full Name:", full_name)
print("Upper Case:", full_name.upper())
print("Lower Case:", full_name.lower())
```

# 🔹 Q13 (Sentence Formatting) User se aik sentence lo aur usay: 1) title case 2)capitalize  3)length print karo

```python
# type your code here
```

```python

sentence = input("Enter a sentence: ")

print("Title Case:", sentence.title())
print("Capitalize:", sentence.capitalize())
print("Length:", len(sentence))
```
